```python
## Task 6
hours = range(1, 9)
passengers = flight.groupby('hours').sum().passengers

plt.plot(hours, passengers)
plt.show()
plt.clf()

sns.lmplot(x='hours', y='passengers', data=flight, x_jitter = .15, y_jitter = .15, fit_reg = False)
plt.show()
```

```python
## Task 7
sns.scatterplot(data=flight, x='coach_price', y='firstclass_price', hue='weekend')
plt.legend()
plt.show()

```

```python
## Task 8
prices = flight.groupby(['redeye', 'day_of_week'])['coach_price'].mean()

prices_red_day = prices.reset_index().pivot(index='redeye', columns='day_of_week', values='coach_price')

sns.heatmap(prices_red_day, cbar_kws={"label": "Average Price"})
plt.show()
```
